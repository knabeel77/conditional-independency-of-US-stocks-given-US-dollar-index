```python
import pandas as pd
project_data = pd.read_csv("https://raw.githubusercontent.com/knabeel77/conditional-independency-of-US-stocks-given-US-dollar-index/main/project_data.csv", index_col=0)
project_data.head()
```

```python
#Returns of individual columns
AAPL_daily_returns =project_data['AAPL'].pct_change()
DX_NYB_daily_returns =project_data['DX-Y.NYB'].pct_change()
HSI_daily_returns =project_data['HSI'].pct_change()
QQQ_daily_returns =project_data['QQQ'].pct_change()
XLE_daily_returns =project_data['XLE'].pct_change()
XLF_daily_returns =project_data['XLF'].pct_change()
XLK_daily_returns =project_data['XLK'].pct_change()
XLP_daily_returns =project_data['XLP'].pct_change()
XLU_daily_returns =project_data['XLU'].pct_change()
XLV_daily_returns =project_data['XLV'].pct_change()
FTSE_daily_returns =project_data['^FTSE'].pct_change()
```

```python
# concating all the column returns into single datafram
data_returns=pd.DataFrame(AAPL_daily_returns)
data_returns.insert(1,"DXY",DX_NYB_daily_returns)
data_returns.insert(2,"HSI",HSI_daily_returns)
data_returns.insert(1,"QQQ",QQQ_daily_returns)
data_returns.insert(2,"XLE",XLE_daily_returns)
data_returns.insert(1,"XLF",XLF_daily_returns)
data_returns.insert(2,"XLK",XLK_daily_returns)
data_returns.insert(2,"XLP",XLP_daily_returns)
data_returns.insert(1,"XLU",XLU_daily_returns)
data_returns.insert(2,"XLV",XLV_daily_returns)
data_returns.insert(2,"FTSE",FTSE_daily_returns)
```
